```python
# Filter the basics table down to only include the US by using the filter akas dataframe.  
# According to project instructions the akas has already been filtered and non us movies have already been removed so this why we used as in (akas
# We used akas 'titleId' and basics 'tconst' to connect them since these both mean the same in both dataframes  
# ou can use panda's .isin() method to remove non-US movies from title basics 
filter_us_titles = basics['tconst'].isin(akas['titleId'])
basics = basics[filter_us_titles]
basics
```

```python
basics = basics.replace({'\\N':np.nan})
```

```python
# Dropping Null values from Drug column
basics = basics.dropna(subset=["runtimeMinutes", "genres"])
```

```python
# Convert startYear to a float dtype
basics['startYear'] = basics['startYear'].astype(float)
```

```python
basics = basics.loc[(basics['startYear'] >= 2000) & (basics['startYear'] <= 2022)]
```

```python
# Eliminate movies that include "Documentary" in genre:
# You will need to create a filter for if the movie's value in the genres column contains the word Documentary.
filter_documentaries = basics['genres'].str.contains('Documentary')
```

```python
# Exclude movies in the documentary category.
# You will also want to use the ~ operator to take the inverse of your Trues/Falses.
basics = basics[~filter_documentaries]
```
